```python
import pandas as pd
import matplotlib.pyplot as plt;
from sklearn import datasets,linear_model,model_selection
```

```python
linnerud_dataset = datasets.load_linnerud();
print(linnerud_dataset.feature_names);
print(linnerud_dataset.target_names);

#To get the label names from plotting on x and y;
y_label_name = linnerud_dataset.target_names[1];
x_label_name = linnerud_dataset.target_names[1];
#Get the  waist the second column from the targets

y= linnerud_dataset.target[:,1].reshape((-1,1));
#Get the  second column(situps) from the data
x= linnerud_dataset.data[:,1].reshape((-1,1))
```

```python
x_train,x_test,y_train,y_test = model_selection.train_test_split(x,y,test_size=0.2,random_state=0);
model = linear_model.LinearRegression();
model.fit(x_train,y_train);
```

```python
y_predict = model.predict(x_test);
```

```python
plt.scatter(x_test,y_test);
plt.plot(x_test,y_predict,color="blue");
plt.xlabel(x_label_name);
plt.ylabel(y_label_name);
plt.show();
```
